```python
# DataFrame
transposed_gmt_df = gmt_df.transpose()

# 假設 transposed_gmt_df 是轉置後的 DataFrame
transposed_gmt_df = transposed_gmt_df.drop(transposed_gmt_df.index[0])

# 列印修改後的 DataFrame
transposed_gmt_df.head()
```

```python
# 新建一個新的欄位用於存取
transposed_gmt_df['category'] = transposed_gmt_df.index

# 假設 transposed_gmt_df 是轉置後的 DataFrame
transposed_gmt_df = transposed_gmt_df.reset_index(drop=True)

# 列印修改後的 DataFrame
print(transposed_gmt_df.head())
```

```python
# 假設 transposed_gmt_df 是轉置後的 DataFrame
column_mapping = {}

for i in range(96):
    quarter = chr(65 + (i % 4))  # 使用 A、B、C、D 來表示每個季度
    month = 1 + (i // 4)  # 使用 1、2、3、4 來表示月
    if month <= 12:
        year = 2022
    else:
        year = 2023
        month -= 12
    new_column_name = f'{year}_{month}{quarter}'
    column_mapping[i] = new_column_name

transposed_gmt_df = transposed_gmt_df.rename(columns=column_mapping)

# 列印修改後的 DataFrame
transposed_gmt_df.head()
```

# 爆品統計

## 總分析

```python
# A時段：01~06
# B時段：07~12
# C時段：13~18
# D時段：19~00

transposed_gmt_df['total_salesAmount'] = transposed_gmt_df.iloc[:, :96].sum(axis=1)

# 創建2022年總銷售金額欄位
transposed_gmt_df['total_salesAmount_2022'] = transposed_gmt_df.iloc[:, 0:48].sum(axis=1)

# 創建2023年總銷售金額欄位
transposed_gmt_df['total_salesAmount_2023'] = transposed_gmt_df.iloc[:, 48:96].sum(axis=1)

transposed_gmt_df.head()
```

## 各時段爆品統計

```python
# 獲取所有欄位名中包含'A'的欄位
columns_with_A = [col for col in transposed_gmt_df.columns if ('A' in col) and ('2023' in col)]

# 獲取所有欄位名中包含'B'的欄位
columns_with_B = [col for col in transposed_gmt_df.columns if ('B' in col) and ('2023' in col)]

# 獲取所有欄位名中包含'C'的欄位
columns_with_C = [col for col in transposed_gmt_df.columns if ('C' in col) and ('2023' in col)]

# 獲取所有欄位名中包含'D'的欄位
columns_with_D = [col for col in transposed_gmt_df.columns if ('D' in col) and ('2023' in col)]

# 將各個類別的A欄位總和，並創建新的欄位'ASum'
transposed_gmt_df['2023_ASum'] = transposed_gmt_df[columns_with_A].sum(axis=1)

# 將各個類別的B欄位總和，並創建新的欄位'BSum'
transposed_gmt_df['2023_BSum'] = transposed_gmt_df[columns_with_B].sum(axis=1)

# 將各個類別的C欄位總和，並創建新的欄位'CSum'
transposed_gmt_df['2023_CSum'] = transposed_gmt_df[columns_with_C].sum(axis=1)

# 將各個類別的D欄位總和，並創建新的欄位'DSum'
transposed_gmt_df['2023_DSum'] = transposed_gmt_df[columns_with_D].sum(axis=1)

# 列印包含新欄位的資料框
transposed_gmt_df[['category', '2023_ASum', '2023_BSum', '2023_CSum', '2023_DSum', 'total_salesAmount_2023']]
```

```python
# 獲取每個類別的A、B、C、D總和欄位
a_sum = transposed_gmt_df['2023_ASum']
b_sum = transposed_gmt_df['2023_BSum']
c_sum = transposed_gmt_df['2023_CSum']
d_sum = transposed_gmt_df['2023_DSum']
total_sales_2023 = transposed_gmt_df['total_salesAmount_2023']


# 對A總和欄位進行排名
a_ranked = a_sum.rank(ascending=False, method='dense').astype(int)

# 對B總和欄位進行排名
b_ranked = b_sum.rank(ascending=False, method='dense').astype(int)

# 對C總和欄位進行排名
c_ranked = c_sum.rank(ascending=False, method='dense').astype(int)

# 對D總和欄位進行排名
d_ranked = d_sum.rank(ascending=False, method='dense').astype(int)

# 對總銷售金額欄位進行排名
total_sales_ranked = total_sales_2023.rank(ascending=False, method='dense').astype(int)


# 創建新的資料框，包含category、ASum、ARank、BSum、BRank、CSum、CRank、DSum、DRank、total_salesAmount欄位
ranked_df = pd.DataFrame({
    'category': transposed_gmt_df['category'],
    'ASum_2023': a_sum,
    'ARank_2023': a_ranked,
    'BSum_2023': b_sum,
    'BRank_2023': b_ranked,
    'CSum_2023': c_sum,
    'CRank_2023': c_ranked,
    'DSum_2023': d_sum,
    'DRank_2023': d_ranked,
    'total_salesAmount_2023': total_sales_2023,
    'total_salesAmountRank_2023': total_sales_ranked
})

# 列印包含排名資訊的資料框
ranked_df

```

```python
import pandas as pd

# 選擇0到95列，計算每列的最高值的索引
max_indices = transposed_gmt_df.iloc[:, 48:96].idxmax()

period_columns = max_indices.index.tolist()

# 使用最高值的索引獲取對應的"category"值
category_values = transposed_gmt_df.loc[max_indices, 'category'].tolist()

salesAmount_values = []
for i, id in enumerate(max_indices.tolist()):
    salesAmount_values.append(transposed_gmt_df.loc[id, max_indices.index[i]])

# 創建包含兩個列表的DataFrame
result_df = pd.DataFrame({'Period_Columns': period_columns, 'Category_Values': category_values, 'Sales_Amount': salesAmount_values})

# 列印結果
result_df
```

```python
import pandas as pd

subset = transposed_gmt_df.iloc[:, 48:96]

# 初始化一個空的 DataFrame 來存儲前三名
top3_df = pd.DataFrame()

# 初始化一個空的列表，用於存儲對應的 category
category_list = []

# 迭代每個欄位，找到前三名的數據
for column in subset.columns:
    # 使用 nlargest 函數找到前三名的索引
    top3_indices = subset[column].nlargest(3).index
    # 使用這些索引獲取相對應的數據
    top3_data = subset.loc[top3_indices, [column]]
    # 為這些數據添加一個欄位，表示欄位名稱
    top3_data['Column_Name'] = column
    # 將這些數據添加到 top3_df 中
    top3_df = pd.concat([top3_df, top3_data])
    # 獲取相應的 category
    category_values = transposed_gmt_df.loc[top3_indices, 'category'].tolist()
    category_list.extend(category_values)

# 將 category_list 添加到 top3_df 中
top3_df['Category'] = category_list

# 重設索引
top3_df.reset_index(drop=True, inplace=True)

# 列印前三名的資料
top3_df
```

```python
top3_df.info()
```

```python
import pandas as pd

# 假設你的 DataFrame 名稱為 top3_df
# 初始化一個空的 DataFrame 來存儲結果
new_df = pd.DataFrame(columns=['Column_Name', 'top3_value_list', 'categories_list'])

# 遍歷唯一的 Column_Name 值
unique_column_names = top3_df['Column_Name'].unique()
for column_name in unique_column_names:
    # 選擇相應的數據，只有三個非 NaN 值
    selected_data = top3_df[top3_df['Column_Name'] == column_name].loc[:, column_name].values
    selected_data = selected_data[~pd.isna(selected_data)].tolist()

    # 選擇相應的 Category 值
    category = top3_df[top3_df['Column_Name'] == column_name]['Category'].values

    # 將數據轉換成字符串形式
    selected_data = [str(item) for item in selected_data]

    # 將數據添加到新的 DataFrame
    new_df = new_df.append({'Column_Name': column_name, 'top3_value_list': selected_data, 'categories_list': category}, ignore_index=True)

# 列印新的 DataFrame
new_df
```

```python
A_list = ['1A', '2A', '3A', '4A', '5A', '6A', '7A', '8A', '9A', '10A', '11A', '12A']
B_list = ['1B', '2B', '3B', '4B', '5B', '6B', '7B', '8B', '9B', '10B', '11B', '12B']
C_list = ['1C', '2C', '3C', '4C', '5C', '6C', '7C', '8C', '9C', '10C', '11C', '12C']
D_list = ['1D', '2D', '3D', '4D', '5D', '6D', '7D', '8D', '9D', '10D', '11D', '12D']

# 創建存儲結果的新列表
A_list_2022 = []
B_list_2022 = []
C_list_2022 = []
D_list_2022 = []

A_list_2023 = []
B_list_2023 = []
C_list_2023 = []
D_list_2023 = []

# 遍歷A_list，將每個值前面添加'2022_'和'2023_'並存儲到新的列表中
for item in A_list:
    A_list_2022.append('2022_' + item)
    A_list_2023.append('2023_' + item)

# 遍歷B_list，將每個值前面添加'2022_'和'2023_'並存儲到新的列表中
for item in B_list:
    B_list_2022.append('2022_' + item)
    B_list_2023.append('2023_' + item)

# 遍歷C_list，將每個值前面添加'2022_'和'2023_'並存儲到新的列表中
for item in C_list:
    C_list_2022.append('2022_' + item)
    C_list_2023.append('2023_' + item)

# 遍歷D_list，將每個值前面添加'2022_'和'2023_'並存儲到新的列表中
for item in D_list:
    D_list_2022.append('2022_' + item)
    D_list_2023.append('2023_' + item)
```

```python
import plotly.express as px
import pandas as pd

transposed_gmt_df['Total_A_2023'] = transposed_gmt_df[A_list_2023].sum(axis=1)
transposed_gmt_df['Total_B_2023'] = transposed_gmt_df[B_list_2023].sum(axis=1)
transposed_gmt_df['Total_C_2023'] = transposed_gmt_df[C_list_2023].sum(axis=1)
transposed_gmt_df['Total_D_2023'] = transposed_gmt_df[D_list_2023].sum(axis=1)

# 創建一個新的 DataFrame 存儲四個欄位的總和
new_df = pd.DataFrame({
    'Interval': ['Total_A_2023', 'Total_B_2023', 'Total_C_2023', 'Total_D_2023'],
    'Total_Sum_2023': [transposed_gmt_df['Total_A_2023'].sum(), transposed_gmt_df['Total_B_2023'].sum(), transposed_gmt_df['Total_C_2023'].sum(), transposed_gmt_df['Total_D_2023'].sum()]
})

# 創建柱狀圖
fig = px.bar(new_df, x='Interval', y='Total_Sum_2023', title='2023 每天四個時段的銷售對比')

# 顯示圖形
fig.show()
```

## 季度總分析

```python
import pandas as pd

# 根據月份將資料分成四個季度，這裡假設每個季度包括3個月
quarters = {
    'Q1_2023': ['1A', '1B', '1C', '1D', '2A', '2B', '2C', '2D', '3A', '3B', '3C', '3D'],
    'Q2_2023': ['4A', '4B', '4C', '4D', '5A', '5B', '5C', '5D', '6A', '6B', '6C', '6D'],
    'Q3_2023': ['7A', '7B', '7C', '7D', '8A', '8B', '8C', '8D', '9A', '9B', '9C', '9D'],
    'Q4_2023': ['10A', '10B', '10C', '10D', '11A', '11B', '11C', '11D', '12A', '12B', '12C', '12D']
}

# 遍歷字典中的每個列表，並將每個值前面添加'2023_'
for key, value in quarters.items():
    quarters[key] = ['2023_' + item for item in value]

# 列印修改後的字典
print(quarters)

# 創建新的 DataFrame 存儲四個季度的總和
quarterly_totals = pd.DataFrame()

for quarter, columns in quarters.items():
    transposed_gmt_df[f'{quarter}_Sum'] = transposed_gmt_df[columns].sum(axis=1)
    quarterly_totals[quarter] = transposed_gmt_df[f'{quarter}_Sum']

# 列印包含季度總和的新 DataFrame
quarterly_totals
```

```python
import pandas as pd
import plotly.express as px

# 計算各個季度的總和，並存儲在相應的變數中
Q1_sum = quarterly_totals['Q1_2023'].sum()
Q2_sum = quarterly_totals['Q2_2023'].sum()
Q3_sum = quarterly_totals['Q3_2023'].sum()
Q4_sum = quarterly_totals['Q4_2023'].sum()

# 創建一個 DataFrame 存儲季度總和
quarterly_totals_stat = pd.DataFrame({
    '季度': ['Q1_2023', 'Q2_2023', 'Q3_2023', 'Q4_2023'],
    '銷售總和': [Q1_sum, Q2_sum, Q3_sum, Q4_sum]
})

# 創建柱狀圖
fig = px.bar(quarterly_totals_stat, x='季度', y='銷售總和', title='2023 每個季度的銷售總和')

# 顯示圖形
fig.show()
```
