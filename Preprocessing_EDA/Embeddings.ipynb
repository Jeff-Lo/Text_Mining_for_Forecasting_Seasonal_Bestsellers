# Data Preprocessing

```python
file_name = '商品資料-2022_all_preprocessed.csv'
file_path = f"{folder_path}/{file_name}"

full_df = pd.read_csv(file_path)
```

```python
import re

unique_names_set = set()

def rename(name):
    name_pattern = re.compile(r'[\u4e00-\u9fa5]+')
    normalized_name = ''.join(re.findall(name_pattern, name))
    return normalized_name

full_df['normalized_name'] = full_df['name'].astype(str).apply(rename)
unique_names = full_df['normalized_name'].dropna().unique()
total_unique_names = len(unique_names)
print(f"總共有 {total_unique_names} 個獨特的normalized_name")

```

```python
import pandas as pd
import ast
import matplotlib.pyplot as plt

def preprocess_data(data: pd.DataFrame):
    original_rows = data.shape[0]  # 獲取刪除NaN值前的列
    data = data.dropna()
    data = data[data['normalized_name'] != '']
    affected_rows = original_rows - data.shape[0]

    # 列印受影響的列
    print(f"受影响的行数：{affected_rows}")

    # 時間戳記解析
    data['time'] = pd.to_datetime(data['time'])

    # 處理商品類別
    data['category'] = data['category'].apply(lambda x: ast.literal_eval(x))

    # 數據類型轉換
    data['total_quantity'] = data['total_quantity'].astype(int)
    data['price'] = data['price'].astype(float)
    data['total_amount'] = data['total_amount'].astype(float)

    # 數據探索
    print(data.head())  # 查看前幾行數據
    print(data.describe())  # 查看數據的統計訊息

    # 可視化，繪製直方圖
    plt.hist(data['price'], bins=20)
    plt.xlabel('Price')
    plt.ylabel('Frequency')
    plt.title('Distribution of Prices')
    plt.show()

    return data
```

```python
full_df = preprocess_data(full_df)
full_df.info()
```

```python
file_name = '商品資料-2022_all_preprocessed.csv'
file_path = f"{folder_path}/{file_name}"

full_df.to_csv(file_path, index=False)
```

```python
from transformers import DistilBertTokenizer, DistilBertModel
import torch

# 初始化DistilBERT tokenizer和model
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
model = DistilBertModel.from_pretrained('distilbert-base-uncased')

# 將模型和數據移動到CUDA設備
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
# 列印出獨特的 category
unique_categories = set()
for index, row in full_df.iterrows():
    if isinstance(row['category'], list):
        unique_categories.update(row['category'])

print("独特的category:")
for category in unique_categories:
    print(category)
```

```python
import json

def Save_Category_embeddings(EmbeddingJson, Name):
    rootEmbedding_path = '/content/drive/Shareddrives/TextMining_Disk/LAB_Folder/Final_project/datasets/embeddings/Categories Embeddings'
    Save_path = f"{rootEmbedding_path}/{Name}"

    with open(Save_path, 'w') as file:
        json.dump(EmbeddingJson, file)

    print(f"Now saving the embedding of category: {category}")
```

```python
# Iterowing the unique_categories and do embeddings
for i, category in enumerate(unique_categories):
    print('---------------------------------------------------')
    print(f"[{i}] Now doing embedding category: {category}...")
    filtered_df = full_df[full_df['category'].apply(lambda x: category in x)]
    unique_names_set = filtered_df['normalized_name'].unique().tolist()

    # 創建一個空字典來存儲嵌入結果
    name_embeddings = {}

    # 遍歷unique_names中的每個名稱，並獲得其嵌入向量
    for name in unique_names_set:
        inputs = tokenizer(name, return_tensors='pt', padding=True, truncation=True).to(device)
        with torch.no_grad():
            outputs = model(**inputs)
        embeddings = outputs.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()
        name_embeddings[name] = embeddings.tolist()
        del outputs, embeddings

    # Save the embedding
    Save_Category_embeddings(name_embeddings, category)

    # move the embedding out of the RAM
    del name_embeddings
```
